```python
df.drop("Avatar",inplace=True,axis=1)
```

```python
X1=df.drop("Yearly Amount Spent",axis=1)
Y1=df["Yearly Amount Spent"]
```

```python
#Implementing Feature Encoding and Scaling
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X1['Email']=le.fit_transform(X1['Email'])
```

```python
from sklearn.model_selection import train_test_split
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1,Y1,random_state=42,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
le=LinearRegression()
```

```python
le.fit(X1_train,Y1_train)
```

```python
Y1_pred=le.predict(X1_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print("R2_score : ",r2_score(Y1_pred,Y1_test))
print("Mean Absolute error : ",mean_absolute_error(Y1_pred,Y1_test))
print("Mean Squared Erroe : ",mean_squared_error(Y1_pred,Y1_test))
```

```python
from sklearn.linear_model import Ridge, Lasso

re = Ridge()
la=Lasso()
```

```python
re.fit(X1_train,Y1_train)
la.fit(X1_train,Y1_train)
```

```python
Y11=re.predict((X1_test))
Y12=la.predict((X1_test))
```

```python
print("R2_score : ",r2_score(Y11,Y1_test))
print("Mean Absolute error : ",mean_absolute_error(Y11,Y1_test))
print("Mean Squared Erroe : ",mean_squared_error(Y11,Y1_test))
```

```python
print("R2_score : ",r2_score(Y12,Y1_test))
print("Mean Absolute error : ",mean_absolute_error(Y12,Y1_test))
print("Mean Squared Erroe : ",mean_squared_error(Y12,Y1_test))
```
